```python
import numpy as np
```

```python
#number of queens
N=8
```

```python
# 8x8 matrix with all elements set to 0
board=np.zeros([8,8],dtype=int)
board=board.tolist()
board
```

```python
# 0 = queen is absent at the place
# 1 = queen is present
```

```python
def attack(i, j):
    #checking vertically
    for r in range(0,N):
        if board[i][r]==1:    #if there is a queen
            return False
        
    #checking horizontally   
    for c in range(0,N):
        if board[c][j]==1:    #if there is a queen
            return False    
        
    #checking diagonally
    for r in range(0,N):
        for c in range(0,N):
            if (r+c==i+j) or (r-c==i-j):
                if board[r][c]==1:
                    return False
    return True
```

```python
def N_queens(board):
    for i in range(0,N):
        for j in range(0,N):
            if attack(i,j) and board[i][j]==0:  #place if there is no queen in given position and if empty, check if we can place a queen
                board[i][j] = 1                 #if we can, then place it
                N_queens(board)                 #pass board for recursive solution
                
                #if there are 8 queens
                if sum(sum(a) for a in board)== N :
                    return board
                
                board[i][j] = 0                 #remove the previous placed queen
                
    return board
```

```python
print(np.matrix(N_queens(board)))
```
